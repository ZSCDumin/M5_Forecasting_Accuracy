```python
import pandas as pd
import numpy as  np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('../data/sales_train_evaluation.csv')

########################### Make Grid
#################################################################################
print('Create Grid')

index_columns = ['id','item_id','dept_id','cat_id','store_id','state_id']
grid_df = pd.melt(sales, 
                  id_vars = index_columns, 
                  var_name = 'd', 
                  value_name = 'sales')

grid_df['day'] = grid_df['d'].map(lambda x: np.int32(x.split('_')[-1]))

grid_df.head()
```

```python
ts_store_sales = grid_df[[ 'store_id', 'day', 'sales']].groupby([ 'store_id', 'day']).agg(sum).reset_index()
ts_store_sales.columns = [ 'store_id', 'day', 'sales']
ts_store_sales.head()
```

```python
#STORES ids
STORES_IDS = grid_df['store_id']
STORES_IDS = list(STORES_IDS.unique())
STORES_IDS
```

```python
store = 'CA_1'
tmp = ts_store_sales[(ts_store_sales.store_id==store) & (ts_store_sales.day>(ts_store_sales.day.max()-28))]['sales'].values
tmp
```

# submission42做趋势

```python
sub_42 = pd.read_csv('../code_47487/submission42_xxxxx_LGBM.csv')
sub_42.head()
```

```python
######################################## day 1-7
trend_value = 1.001
for i in range(1, 8):
    col ='F'+str(i)
    print('=====================> col = {col}, trend_value = {trend_value} '.format(col=col, trend_value=trend_value))
    print('change before: ', sub_42[col].sum())
    sub_42[col] *= trend_value
    print('change after: ', sub_42[col].sum())
    if i<=3:
        trend_value += 0.0001
    else:
        trend_value += 0.0002
```
